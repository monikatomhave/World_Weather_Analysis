```python
preferred_country_df =vacation_df.loc[(vacation_df["Country"] == country)]
preferred_country_df.head(10)
```

```python
city_list= []
for c in preferred_country_df["City"]:
    if c not in city_list:
        city_list.append(c)
print(city_list)
start_city = (input("What city you would like to start your trip in? "))
first_stop = (input("What city you would like to visit second in your trip? "))
second_stop = (input("What city you would like to visit third in your trip? "))
third_stop = (input("What city you would like to visit fourth in your trip? "))
```

```python
# 2. Using the template add the city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Current Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# 3a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in preferred_country_df.iterrows()]

# 3b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = vacation_df[["Lat", "Lng"]]
```

```python
# 4a. Add a marker layer for each city to the map.
locations = preferred_country_df[["Lat", "Lng"]]
max_temp = preferred_country_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
# 4b. Display the figure
fig
```

```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 
# 5. Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city.

vacation_start = vacation_df.loc[vacation_df["City"] == start_city]
vacation_end = vacation_df.loc[vacation_df["City"] == start_city]
vacation_stop1 = vacation_df.loc[vacation_df["City"] == first_stop]
vacation_stop2 = vacation_df.loc[vacation_df["City"] == second_stop] 
vacation_stop3 = vacation_df.loc[vacation_df["City"] == third_stop] 

print(vacation_start, vacation_stop1, vacation_stop2, vacation_stop3, vacation_end)
```

```python
# 6. Get the latitude-longitude pairs as tuples from each city DataFrame using the to_numpy function and list indexing.
start = vacation_start[["Lat", "Lng"]].to_numpy()[0]
end = vacation_end[["Lat", "Lng"]].to_numpy()[0]
stop1 = vacation_stop1[["Lat", "Lng"]].to_numpy()[0]
stop2 = vacation_stop2[["Lat", "Lng"]].to_numpy()[0]
stop3 = vacation_stop3[["Lat", "Lng"]].to_numpy()[0]

print(start, stop1, stop2, stop3, end)
```

```python
# 7. Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, and stop3 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING".

fig = gmaps.figure()

#start_stop1 = gmaps.directions_layer(start, stop1)
#fig.add_layer(start_stop1)
#stop1_stop2 = gmaps.directions_layer(stop1, stop2)
#fig.add_layer(stop1_stop2)
#stop2_stop3 = gmaps.directions_layer(stop2, stop3)
#fig.add_layer(stop2_stop3)
#stop3_end = gmaps.directions_layer(stop3, end)
#fig.add_layer(stop3_end)

start2end = gmaps.directions_layer(start, end, waypoints = [stop1, stop2, stop3], travel_mode = "DRIVING")
fig.add_layer(start2end)

fig
```

```python
# 8. To create a marker layer map between the four cities.
#  Combine the four city DataFrames into one DataFrame using the concat() function.
itinerary_df = pd.concat([vacation_start, vacation_stop1, vacation_stop2, vacation_stop3],ignore_index=True)
itinerary_df
```

```python
# 9 Using the template add city name, the country code, the weather description and maximum temperature for the city. 
info_box_template = """
<dl>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Weather Description</dt><dd>{Current Description}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# 10a Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]

# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = itinerary_df[["Lat", "Lng"]]
```

```python
# 11a. Add a marker layer for each city to the map.
locations = itinerary_df[["Lat", "Lng"]]
max_temp = itinerary_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
# 11b. Display the figure
fig
```
