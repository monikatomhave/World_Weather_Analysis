```python
# 6.5.4 cont'

# Import Dependancies

import requests
import gmaps

# Import API key

from config import g_key
```

```python
# set params to search fro a hotel in Paris

params = {
    "radius": 5000,
    "types": "lodging", 
    "key": g_key,
    "location": "48.8566, 2.3522"}

# use the base URL to search for hotels in Paris
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Make a request and get the JSON data from the search
hotels = requests.get(base_url, params = params).json()

hotels
```
